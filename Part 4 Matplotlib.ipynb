```python
fig , axes = plt.subplots(nrows = 3, ncols = 1 )


# axes[0].set_title('First Plot')

# axes[1].set_title('Second Plot')
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[2].plot(y,x)


# If you try and print out your axes objects in both cases, you will notice that it is a numpy array, containing matplotlib.axes._subplots.AxesSubplot objects.
# These are what you access to plot on them (in case of multiple rows, or columns (not both) using axes[0], axes[1] and so on).
# When you have a multiple of either rows or columns (not both) this array is 1-dimensional, so you can access individual values with one index (axes[0]).
# However when you have multiple rows AND multiple columns, this numpy array is 2-dimensional and you need to index in both dimensions,
# for example axes[0][0] or axes[0,0] for the very first axis. 
# That's why your error says numpy.ndarray object has no attribute plot, because you are not accessing an axis, but a numpy.ndarray within numpy.ndarray
```
